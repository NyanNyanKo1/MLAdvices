## 1 Кластерный анализ


Для кластеризации данных из предоставленного датафрейма, мы можем использовать различные алгоритмы кластеризации. В данном случае, я предложу применить три популярных метода кластеризации: K-Means, Agglomerative Hierarchical Clustering и DBSCAN.

1. K-Means:
  K-Means - это итеративный алгоритм кластеризации, который разделяет данные на K кластеров. Он пытается минимизировать сумму квадратов расстояний между точками и центроидами кластеров.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data2.csv')

# Выбираем признаки для кластеризации
selected_features = ['start_use_datetime_year', 'start_use_datetime_month', 'start_use_datetime_day',
                     'end_use_datetime_year', 'end_use_datetime_month', 'end_use_datetime_day',
                     'Total_time_using_app_seconds']

# Извлекаем данные для кластеризации
data_for_clustering = df[selected_features]

# Нормализуем данные для обеспечения одинаковой важности каждого признака
scaler = StandardScaler()
data_for_clustering_scaled = scaler.fit_transform(data_for_clustering)

# Применяем K-Means с числом кластеров K=3 и указываем значение n_init=10
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(data_for_clustering_scaled)
```

2. Agglomerative Hierarchical Clustering:
Агломеративная иерархическая кластеризация начинается с каждой точки в отдельном кластере и последовательно объединяет ближайшие кластеры до тех пор, пока не будет достигнуто заданное количество кластеров.

```python
from sklearn.cluster import AgglomerativeClustering

# Применяем Agglomerative Hierarchical Clustering с числом кластеров K=3
agg_clustering = AgglomerativeClustering(n_clusters=3)
df['agg_cluster'] = agg_clustering.fit_predict(data_for_clustering_scaled)
```

3. DBSCAN:
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) - это алгоритм кластеризации, который находит области с высокой плотностью точек, разделяя их от областей с низкой плотностью.

```python
from sklearn.cluster import DBSCAN

# Применяем DBSCAN с параметрами epsilon=0.5 и min_samples=5
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(data_for_clustering_scaled)
```

Обоснование выбора алгоритмов:

* K-Means: Хорошо работает, когда данные имеют компактные кластеры и примерно одинаковую дисперсию.
* Agglomerative Hierarchical Clustering: Подходит для иерархической структуры кластеров, когда точки могут объединяться в подгруппы.
* DBSCAN: Эффективен для обнаружения выбросов и работы с кластерами произвольной формы.


Для оценки качества кластеризации, нам нужно выбрать соответствующую метрику, которая поможет оценить, насколько хорошо объекты внутри кластеров схожи между собой и насколько отличаются объекты между разными кластерами.

Одной из самых распространенных метрик для оценки качества кластеризации является Silhouette Score (коэффициент силуэта). Silhouette Score вычисляется для каждого объекта и предоставляет оценку "плотности" кластеризации. Он варьируется от -1 до 1, где значения ближе к 1 указывают на хорошую кластеризацию, а значения ближе к -1 указывают на неправильную кластеризацию.

Обоснование выбора Silhouette Score:

Интерпретируемость:
1. Silhouette Score легко интерпретировать. Чем ближе значение к 1, тем лучше кластеризация.
2. Универсальность: Silhouette Score применим для различных алгоритмов кластеризации и типов данных.
3. Не требует информации о "правильных" метках: Silhouette Score не требует знания истинных меток кластеров, что делает его полезным для задач без учителя.
Как использовать Silhouette Score в Python:

```python
from sklearn.metrics import silhouette_score

# Для K-Means
kmeans_silhouette = silhouette_score(data_for_clustering_scaled, df['kmeans_cluster'])
print("K-Means Silhouette Score:", kmeans_silhouette)

# Для Agglomerative Hierarchical Clustering
agg_silhouette = silhouette_score(data_for_clustering_scaled, df['agg_cluster'])
print("Agglomerative Silhouette Score:", agg_silhouette)

# Применяем DBSCAN с другими значениями параметров
dbscan = DBSCAN(eps=0.3, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(data_for_clustering_scaled)
```


Для визуального анализа оценки качества кластеризации, мы можем использовать различные графические методы, которые помогут нам понять, как хорошо алгоритмы кластеризации разбили данные на кластеры. Возможным примером является визуализация с использованием разноцветных точек для каждого кластера и центроидов (в случае K-Means)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# Функция для визуализации результатов кластеризации
def plot_clusters(data, cluster_labels, title):
    # Снижаем размерность данных до 2D с помощью t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    data_tsne = tsne.fit_transform(data)

    # Создаем датафрейм с данными и метками кластеров
    df_tsne = pd.DataFrame({'Dimension 1': data_tsne[:, 0], 'Dimension 2': data_tsne[:, 1], 'Cluster': cluster_labels})

    # Визуализируем данные
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='Dimension 1', y='Dimension 2', hue='Cluster', data=df_tsne, palette='tab10', s=100, alpha=0.7)
    plt.title(title)
    plt.show()

# Визуализация результатов K-Means
plot_clusters(data_for_clustering_scaled, df['kmeans_cluster'], 'K-Means Clustering')

# Визуализация результатов Agglomerative Hierarchical Clustering
plot_clusters(data_for_clustering_scaled, df['agg_cluster'], 'Agglomerative Clustering')

# Визуализация результатов DBSCAN
plot_clusters(data_for_clustering_scaled, df['dbscan_cluster'], 'DBSCAN Clustering')
```

Для разделения данных на обучающую и тестовую выборки с учетом несбалансированности, мы можем использовать стратифицированное разбиение. Стратифицированное разбиение гарантирует, что соотношение классов в обучающей и тестовой выборках будет приближенным к исходному соотношению классов в исходных данных. Это особенно полезно, когда имеются дисбалансы между классами.

Мы будем использовать библиотеку scikit-learn, которая предоставляет инструменты для стратифицированного разбиения данных. В данном примере, предположим, что у нас есть столбец target в датафрейме df, который содержит метки классов.

Основание для использования стратифицированного разбиения данных при кластеризации состоит в том, чтобы сохранить баланс между классами или кластерами в обучающей и тестовой выборках. Это особенно важно, когда в исходных данных есть дисбаланс между классами или кластерами.

Сохранение пропорций: Разбивая данные на обучающую и тестовую выборки, мы хотим сохранить пропорции классов или кластеров в исходных данных. Если мы случайным образом разделим данные, то существует вероятность, что один класс или кластер окажется сильно перерепрезентированным или недостаточно представленным в обучающей или тестовой выборке.

Обобщаемость: Если модель обучается на несбалансированных данных, она может склоняться к классу или кластеру с большим числом представителей, что может привести к смещению и низкой обобщаемости на новых данных.

Адекватность оценки качества: Если в обучающей выборке присутствуют только представители одного класса или кластера, то метрики оценки качества, такие как точность, могут давать завышенные результаты, что может быть обманчиво.

Повышение устойчивости: При стратифицированном разбиении мы повышаем устойчивость качества модели, так как обучение и тестирование происходят на разнообразных данных.

Следовательно, стратифицированное разбиение помогает создать более адекватные и обобщающие модели, учитывая особенности данных и их дисбаланс. Это основание применяется для разделения данных на обучающую и тестовую выборки в контексте кластеризации и других задач машинного обучения.

##3 Построение модели прогнозирования временного ряда

Для прогнозирования временного ряда в данном случае, где у нас есть временные отметки start_use_datetime и end_use_datetime, можно использовать следующие три алгоритма:

ARIMA (Autoregressive Integrated Moving Average):
ARIMA - это статистический метод для анализа и прогнозирования временных рядов. Он основан на авторегрессии, интеграции и скользящем среднем. ARIMA модель хорошо подходит для стационарных временных рядов.

Prophet:
Prophet - это библиотека для прогнозирования временных рядов, разработанная Facebook. Она автоматически обрабатывает недостающие данные, а также учитывает сезонные и праздничные эффекты, что делает ее простой в использовании и эффективной для прогнозирования различных типов временных рядов.

LSTM (Long Short-Term Memory):
LSTM - это тип рекуррентных нейронных сетей, который хорошо подходит для анализа последовательных данных, таких как временные ряды. LSTM способна улавливать зависимости в данных на разных временных шкалах, что делает ее мощным инструментом для прогнозирования временных рядов.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from fbprophet import Prophet
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Загружаем датафрейм
df = pd.read_csv('data2.csv')

# Создаем временной ряд из данных
time_series = df[['start_use_datetime', 'Total_time_using_app_seconds']].copy()
time_series['start_use_datetime'] = pd.to_datetime(time_series['start_use_datetime'])
time_series = time_series.set_index('start_use_datetime')

# ARIMA
arima_model = ARIMA(time_series['Total_time_using_app_seconds'], order=(5,1,0))
arima_fit = arima_model.fit()
arima_predictions = arima_fit.forecast(steps=30)  # Прогнозируем на 30 шагов вперед

# Prophet
prophet_model = Prophet()
prophet_model.fit(time_series.reset_index().rename(columns={'start_use_datetime':'ds', 'Total_time_using_app_seconds':'y'}))
future = prophet_model.make_future_dataframe(periods=30)  # Прогнозируем на 30 шагов вперед
prophet_predictions = prophet_model.predict(future)['yhat'].tail(30)

# LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series)

X_train, y_train = [], []
for i in range(60, len(time_series)):
    X_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1],1)))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(1))

lstm_model.compile(loss='mean_squared_error', optimizer='adam')
lstm_model.fit(X_train, y_train, epochs=100, batch_size=32)

# Прогнозируем на 30 шагов вперед
test_inputs = scaled_data[len(scaled_data) - len(X_train) - 60:]
test_inputs = test_inputs.reshape(-1,1)
test_inputs = scaler.transform(test_inputs)

X_test = []
for i in range(60, 60 + 30):
    X_test.append(test_inputs[i-60:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

lstm_predictions = lstm_model.predict(X_test)
lstm_predictions = scaler.inverse_transform(lstm_predictions)

# lstm_predictions содержит прогнозы на 30 шагов вперед

```

Чтобы определить модель с наилучшим качеством, необходимо выбрать подходящую метрику для оценки временных рядов, такую как среднеквадратичная ошибка (MSE), средняя абсолютная ошибка (MAE)

```python
from sklearn.metrics import mean_squared_error

# Оригинальные временные ряды
original_series = time_series['Total_time_using_app_seconds']

# Оценка качества прогнозов ARIMA
arima_mse = mean_squared_error(original_series[-30:], arima_predictions)
print("MSE ARIMA:", arima_mse)

# Оценка качества прогнозов Prophet
prophet_mse = mean_squared_error(original_series[-30:], prophet_predictions)
print("MSE Prophet:", prophet_mse)

# Оценка качества прогнозов LSTM
lstm_mse = mean_squared_error(original_series[-30:], lstm_predictions)
print("MSE LSTM:", lstm_mse)
```
