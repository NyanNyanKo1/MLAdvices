```python
import csv
import random
from faker import Faker

# Используем библиотеку Faker для генерации рандомных данных
fake = Faker()

# Генерируем данные для таблицы User
def generate_users(num_users):
    users = []
    for _ in range(num_users):
        users.append({
            'id': fake.uuid4(),
            'name': fake.name(),
            'country': fake.country(),
            'language': fake.language_code(),
            'device': fake.word(ext_word_list=['Phone', 'Tablet', 'Laptop', 'Desktop', 'Smartwatch'])
        })
    return users

# Генерируем данные для таблицы App
def generate_apps(num_apps):
    apps = []
    for _ in range(num_apps):
        apps.append({
            'id': fake.uuid4(),
            'name': fake.company()
        })
    return apps

# Генерируем данные для таблицы Use
def generate_use(num_uses, users, apps):
    uses = []
    for _ in range(num_uses):
        user_id = random.choice(users)['id']
        app_id = random.choice(apps)['id']
        start_use_datetime = fake.date_time_this_decade()
        end_use_datetime = fake.date_time_between(start_use_datetime, '+1y')
        uses.append({
            'user_id': user_id,
            'app_id': app_id,
            'start_use_datetime': start_use_datetime.strftime('%Y-%m-%d %H:%M:%S'),
            'end_use_datetime': end_use_datetime.strftime('%Y-%m-%d %H:%M:%S')
        })
    return uses

# Количество записей для каждой таблицы
num_users = 100
num_apps = 50
num_uses = 200

# Генерация данных
users_data = generate_users(num_users)
apps_data = generate_apps(num_apps)
use_data = generate_use(num_uses, users_data, apps_data)

# Сохраняем данные в CSV-файлах
def save_to_csv(data, filename, fieldnames):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

save_to_csv(users_data, 'users.csv', fieldnames=['id', 'name', 'country', 'language', 'device'])
save_to_csv(apps_data, 'apps.csv', fieldnames=['id', 'name'])
save_to_csv(use_data, 'use.csv', fieldnames=['user_id', 'app_id', 'start_use_datetime', 'end_use_datetime'])
```

Для расчета маркетинговых показателей эффективности, мы можем создать функции, которые будут анализировать данные в DataFrame merged_df и вычислять нужные показатели. Вот несколько примеров функций для расчета маркетинговых показателей:

Среднее время использования приложения по странам:
Этот показатель поможет определить, в каких странах пользователи проводят больше времени, что может быть полезно для определения целевой аудитории или для оценки успешности маркетинговых кампаний в определенных странах.

```python
def average_time_by_country(df):
    avg_time_by_country = df.groupby('country')['Total_time_using_app'].mean()
    return avg_time_by_country
```

Среднее время использования приложения по устройствам:
Этот показатель поможет понять, на каких устройствах пользователи чаще всего используют приложение, что может помочь в оптимизации приложения для конкретных устройств.

```python
def average_time_by_device(df):
    avg_time_by_device = df.groupby('device')['Total_time_using_app'].mean()
    return avg_time_by_device
```

Количество уникальных пользователей по приложениям:
Этот показатель поможет определить популярность каждого приложения и его привлекательность для пользователей.

```python
def unique_users_by_app(df):
    unique_users_by_app = df.groupby('app_id')['user_id'].nunique()
    return unique_users_by_app
```

Среднее время использования приложения в разрезе языков:
Этот показатель поможет понять, насколько успешно приложение адаптировано под различные языки и культуры.

```python
def average_time_by_language(df):
    avg_time_by_language = df.groupby('language')['Total_time_using_app'].mean()
    return avg_time_by_language
```

Количество использований приложения по дням:
Этот показатель поможет выявить тренды использования приложения и определить пиковые периоды активности пользователей.

```python
def app_usage_by_day(df):
    df['use_date'] = df['start_use_datetime'].dt.date
    app_usage_by_day = df.groupby('use_date').size()
    return app_usage_by_day
```

```python
def average_time_per_user(df):
    avg_time_per_user = df.groupby('user_id')['Total_time_using_app'].mean()
    return avg_time_per_user
```

Популярность устройства среди пользователей:
Этот показатель позволяет определить, на каких устройствах пользователи чаще всего используют приложение. Например, это может быть полезно для оптимизации приложения для популярных устройств.

```python
def popular_devices(df):
    popular_devices = df['device'].value_counts()
    return popular_devices
```

Предпочитаемый язык пользователей:
Этот показатель позволяет определить, какой язык используют пользователи, и может быть полезным для адаптации приложения под нужды пользователей разных языковых групп.

```python
def preferred_language(df):
    preferred_language = df['language'].mode().iloc[0]
    return preferred_language
```

Средний интервал между использованиями приложения:
Этот показатель позволяет определить, как часто пользователи возвращаются к использованию приложения.

```python
def average_time_between_uses(df):
    df.sort_values(by=['user_id', 'start_use_datetime'], inplace=True)
    df['time_between_uses'] = df.groupby('user_id')['start_use_datetime'].diff()
    avg_time_between_uses = df.groupby('user_id')['time_between_uses'].mean()
    return avg_time_between_uses
```

Количество использований приложения по дням недели:
Этот показатель поможет понять, в какие дни недели пользователи чаще всего используют приложение.

```python
def app_usage_by_weekday(df):
    df['weekday'] = df['start_use_datetime'].dt.weekday
    app_usage_by_weekday = df.groupby('weekday').size()
    return app_usage_by_weekday
```


Для проверки временного ряда на стационарность, можно использовать статистические тесты, такие как Augmented Dickey-Fuller (ADF) тест. ADF тест используется для проверки единичных корней во временном ряде, что является признаком нестационарности.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller


time_series = df["Total_time_using_app_seconds"]

# Выполняем ADF тест
result = adfuller(time_series)

# Результаты теста
print("ADF статистика:", result[0])
print("p-значение:", result[1])
print("Критические значения:")
for key, value in result[4].items():
    print(f"\t{key}: {value}")

# Оценка стационарности на основе p-значения
if result[1] <= 0.05:
    print("Временной ряд стационарен (отвергаем гипотезу о нестационарности)")
else:
    print("Временной ряд нестационарен (не отвергаем гипотезу о нестационарности)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss


time_series = df['Total_time_using_app_seconds']

# Удалим возможные пропущенные значения в ряду
time_series = time_series.dropna()

# Создадим функцию для проведения теста Дики-Фуллера и вывода результатов
def adf_test(time_series):
    result = adfuller(time_series)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')

# Создадим функцию для проведения теста КПСС и вывода результатов
def kpss_test(time_series):
    result = kpss(time_series)
    print('KPSS Statistic:', result[0])
    print('p-value:', result[1])
    print('Lags Used:', result[2])
    print('Critical Values:')
    for key, value in result[3].items():
        print(f'\t{key}: {value}')
```

```python
# Применяем тест Дики-Фуллера
print("Результаты теста Дики-Фуллера:")
adf_test(time_series)

# Применяем тест КПСС
print("\nРезультаты теста КПСС:")
kpss_test(time_series)
```
