```python
import pandas as pd
```

Преобразуем в датавреймы

```python
df_users = pd.DataFrame(users)
```

```python
df_users.set_index("id")
```

```python
df_apps = pd.DataFrame(apps)
```

```python
df_apps.set_index("id")
```

```python
df_stats = pd.DataFrame(stats)
```

Преобразуем поле с датой/временем в дату



```python
# если загружаем с .csv

# data = pd.read_csv('your_dataset.csv', parse_dates=['date_begin',"date_end" ])

# если уже есть DataFrame

df_stats['date_begin'] = pd.to_datetime(df_stats['date_begin'])
df_stats['date_end'] = pd.to_datetime(df_stats['date_end'])
```

```python
df_stats
```

Далее выделим из даты отдельно день, месяц, год и время

Для удобства завернем все в цикл

```python
labels = ["date_begin", "date_end"]

for label in labels:
  df_stats[label+'_year'] = pd.to_datetime(df_stats[label]).dt.year
  df_stats[label+'_month'] = pd.to_datetime(df_stats[label]).dt.month
  df_stats[label+'_day'] = pd.to_datetime(df_stats[label]).dt.day
```

```python
df_stats
```

### Объединяем данные

```python
merged_df = pd.merge(df_stats, df_users, left_on='user_id', right_on='id')
```

```python
merged_df
```

Удалим id пользователей, вряд ли будут дальше нужны

```python
merged_df.drop('user_id', axis=1, inplace=True)
```
